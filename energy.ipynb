```python
from sklearn.linear_model import LinearRegression
import seaborn as sns 
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,r2_score

scaler = MinMaxScaler()

x = pd.DataFrame(data['T2'])
y = data['T6']
model = LinearRegression()
model.fit(x,y)
y_pred = model.predict(x)

r2 = r2_score(y_pred,y)
round(r2,2)
```

```python
scaler = MinMaxScaler()
df = data.drop(['date','lights'],axis = 1)
normalized_df = pd.DataFrame(scaler.fit_transform(df),columns = df.columns)
features = normalized_df.drop(columns=['Appliances'])
target =normalized_df['Appliances']
```

```python
x_train,x_test,y_train,y_test = train_test_split(normalized_df,target,test_size=0.3,random_state=42)

model = LinearRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
mae = mean_absolute_error(y_test,y_pred)
round(mae,2)
```

```python
mse =np.sqrt(mean_squared_error(y_test,y_pred))
round(mse,2)
```

```python
r2 = r2_score(y_test,y_pred)
round(r2,2)
```

```python
weight = pd.Series(model.coef_[1:],features.columns).sort_values()
weight_df = pd.DataFrame(weight).reset_index()
weight_df
```

```python
ridge = Ridge(alpha = 0.4)
ridge.fit(x_train,y_train)

y_pred = ridge.predict(x_test)
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
round(rmse,2)
```

```python
ridge2 = Ridge(alpha = 0.001)
ridge2.fit(x_train,y_train)
```

```python
y_pred = ridge2.predict(x_test)
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
round(rmse,3)
```
